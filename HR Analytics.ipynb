```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler as scale
```

```python
data = pd.read_csv('~/Downloads/HR-Employee-Attrition.csv')
data
```

# redundant columns to be deleted

```python
columns_delete = ['EmployeeCount', 'StandardHours', 'Over18']

cleaned_data = data.drop(columns = columns_delete)
```

```python
cleaned_data.info()
```

# Renaming the columns

```python
new_column_names = {'Department' : 'Dept.',
                    'DistanceFromHome' : 'HomeDistance',
                    'OverTime' : 'OT',
                    'Education' : 'Edu.',
                    'EducationField' : 'Edu.Field'}

cleaned_data.rename(columns = new_column_names, inplace = True)
                    
```

```python
cleaned_data.info()
```

# Dropping Duplicates

```python
cleaned_data = cleaned_data.drop_duplicates()
```

# Cleaning Individual Columns

```python
cleaned_data['Attrition'] = cleaned_data['Attrition'].str.strip()
cleaned_data['BusinessTravel'] = cleaned_data['BusinessTravel'].str.strip()        
cleaned_data['Dept.'] = cleaned_data['Dept.'].str.strip()
cleaned_data['JobRole'] = cleaned_data['JobRole'].str.strip()
cleaned_data['OT'] = cleaned_data['OT'].str.strip()
cleaned_data['Gender'] = cleaned_data['Gender'].str.strip()
cleaned_data['Edu.Field'] = cleaned_data['Edu.Field'].str.strip()
            
```

# Remove the NaN values from the dataset

```python
cleaned_data.dropna()
```

# Check for some more Transformations



```python
# Convert categorical data to Numerical
cleaned_data['Attrition'] = cleaned_data['Attrition'].apply(lambda x : 1 if x == "Yes" else 0)
```

```python
cleaned_data
```

```python
# one-hot encoding to object type data
obj_col = ['BusinessTravel', 'Dept.', 'JobRole', 'OT', 'Gender', 'Edu.Field', 'MaritalStatus']

cleaned_data = pd.get_dummies(cleaned_data, columns = obj_col)
```

```python
cleaned_data
```

```python
scaler = scale()
num_col = ['Age','DailyRate','HomeDistance','HourlyRate','MonthlyIncome','MonthlyRate',
           'NumCompaniesWorked','PercentSalaryHike','TotalWorkingYears','YearsAtCompany',
           'YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']
cleaned_data[num_col] = scaler.fit_transform(cleaned_data[num_col])
```

```python
print(cleaned_data)
```

### Plot a correlation map for all numeric variables

```python
correl = cleaned_data.corr()
```

```python
correl
```

```python
plt.figure(figsize=(25, 15))
sns.heatmap(correl, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
